# Exercise 1

```julia
function unconditional_variance(A, Σ)
    old_S = ones(2, 2)
    S = A * old_S * A' + Σ * Σ'
    while sum(abs.(S - old_S)) > 0.0001
        old_S = S
        S = A * old_S * A' + Σ * Σ'
        println(S - old_S)
    end
    return S
end
```

```julia
A = [0.8 -0.2; -0.1 0.7]
Σ = [0.5 0.4; 0.4 0.6]
res = unconditional_variance(A, Σ)
```

```julia
A * res * A' + Σ * Σ'
```

# Exercise 2

```julia
using Plots

function stochastic(γ, θ, y_0, σ, T)
    y = zeros(T + 1)
    y[1] = y_0
    for t in 1:T
        w = randn()
        y[t + 1] = γ + θ * y[t] + σ * w
    end
    return y
end
```

```julia
γ = 1
θ = [0.8, 0.9, 0.98]
y_0 = 0
σ = 1
T = 10000
y1 = stochastic(γ, θ[1], y_0, σ, T)
y2 = stochastic(γ, θ[2], y_0, σ, T)
y3 = stochastic(γ, θ[3], y_0, σ, T)
plot(y1)
plot!(y2)
plot!(y3)
```
